```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
Income=pd.read_csv("data/canada_per_capita_income.csv")
```

```python
Income.head()
```

```python
Income.shape
```

```python
print(Income.columns)
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.title("Per capita Income according per year")
plt.scatter(Income['year'],Income['per capita income (US$)'],color="red", marker="+")
plt.show()
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(Income[['year']],Income['per capita income (US$)'])
```

```python
reg.predict([[2025]])
```
